```python
import pandas as pd
```

```python
dt = pd.read_csv('/Store_CA.csv')
```

```python
dt.info()
```

```python
x = dt.iloc[:,4:]
x.head()
```

```python
from sklearn.cluster import KMeans
```

```python
MonthlySalesRevenue = []
for i in range(1,14):
  kmeans = KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(x)
  MonthlySalesRevenue.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,14),MonthlySalesRevenue)
plt.title('The retail store')
plt.xlabel('cluster')
plt.ylabel('MonthlySalesRevenue')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3,init='k-means++',random_state=42)
kmeans.fit(x)
```

```python
labels = kmeans.labels_
dt['cluster'] = labels
```

```python
plt.scatter(dt['EmployeeEfficiency'],dt['MonthlySalesRevenue'],c=dt['cluster'])
plt.title('Retail Store')
plt.xlabel('EmployeeEfficiency')
plt.ylabel('MonthlySalesRevenue')
plt.show()
```
